```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from your CSV file
data = pd.read_csv("minard_chard/data_from_minard.csv") 
print(data)
```

```python
# Create the scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(data["long"], data["lat"], marker="o", color="blue")

# Add labels and title
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("City Locations")

# Annotate cities with their names
for row in data.iterrows():
    plt.annotate(row[1]["city"], (row[1]["long"], row[1]["lat"]), textcoords="offset points", xytext=(0,10), ha='center')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the data from your CSV file
data = pd.read_csv("minard_chard/data_from_minard.csv")  # Replace with the actual path

# Create the line chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(data["x"], data["y"])

# Create the bar chart using Seaborn
sns.barplot(
    x="x",
    y="size",
    legend=False,
    color="grey",
    width=0.2,
    bottom=data["y"],
    ax=plt.gca(),
)

# Create the color gradient
plt.imshow(
    data["temperature"].values.reshape(100, 100),
    cmap="coolwarm",
    alpha=0.5,
    extent=(0, 1200, 0, 1000),
)

# Add labels and title
plt.xlabel("Distance (km)")
plt.ylabel("Number of soldiers")
plt.title("The Minard Chart")

# Save the chart as a PNG image
plt.savefig("minard_chart.png")

# Create the interactive chart using Plotly Express
fig = px.line(
    data,
    x="x",
    y="y",
    color="temperature",
    width=1000,
    height=500,
    title="The Minard Chart",
)
fig.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the data from your CSV file
data = pd.read_csv("minard_chard/data_from_minard.csv")  # Replace with the actual path

# Convert the DataFrame to a GeoDataFrame
geometry = gpd.points_from_xy(data["long"], data["lat"])
gdf = gpd.GeoDataFrame(data, geometry=geometry)

# Create the base map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Define the extent for the plot (Zoomed-in Europe to Russia)
extent = (20, 50, 40, 70)  # (minx, maxx, miny, maxy)

# Read the cities shapefile
cities = gpd.read_file('cities_shapefile.shp')  # Replace with the actual path

# Plot the map with country boundaries
fig, ax = plt.subplots(figsize=(10, 6))
world.boundary.plot(ax=ax)

# Plot the cities from the shapefile
cities.plot(ax=ax, marker='o', color='red', markersize=50)

# Set the extent
ax.set_xlim(extent[0], extent[1])
ax.set_ylim(extent[2], extent[3])

# Annotate cities with their names
for x, y, label in zip(gdf.geometry.x, gdf.geometry.y, gdf["city"]):
    ax.text(x, y, label, fontsize=8, ha='right')

# Add labels and title
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("City Locations (Zoomed-in) from Europe to Russia using GeoPandas")

# Show the plot
plt.show()
```

```python
import geopandas as gpd
import plotly.express as px

# Load the data from your CSV file
data = pd.read_csv("minard_chard/data_from_minard.csv")  # Replace with the actual path

# Convert the DataFrame to a GeoDataFrame
geometry = gpd.points_from_xy(data["long"], data["lat"])
gdf = gpd.GeoDataFrame(data, geometry=geometry)

# Define the extent for the plot (Zoomed-in Europe to Russia)
extent = (20, 50, 40, 70)  # (minx, maxx, miny, maxy)

# Create a Plotly Express scatter map
fig = px.scatter_geo(
    data_frame=gdf,
    lon="long",
    lat="lat",
    text="city",
    title="City Locations from Europe to Russia using Plotly Express",
    scope="europe",
    color_discrete_sequence=["blue"],  # Set marker color
)

# Show the plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
troops = pd.read_table("minard_chard/troops.txt", header=True)
cities = pd.read_table("minard_chard/cities.txt", header=True)
temps = pd.read_table("minard_chard/temps.txt", header=True, parse_dates=["date"])

# Create the first plot (Minard chart)
fig, ax = plt.subplots(figsize=(16, 4))

for _, row in troops.iterrows():
    ax.plot(row['long'], row['lat'], marker='o', markersize=row['survivors'] / 1000, color='grey' if row['direction'] == 'A' else 'red')

ax.scatter(cities['long'], cities['lat'], color='black')
for _, row in cities.iterrows():
    ax.text(row['long'], row['lat'], row['city'], ha='right', va='top', fontsize=8)

ax.set_xlim(24, 39)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Napoleon\'s March Minard Chart')

plt.savefig("march.pdf")

# Create the second plot (Temperature chart)
plt.figure(figsize=(16, 4))
plt.plot(temps['long'], temps['temp'])
for _, row in temps.iterrows():
    plt.text(row['long'], row['temp'], f"{row['day']} {row['month']}", va='top', fontsize=8)
plt.xlim(24, 39)
plt.xlabel('Longitude')
plt.ylabel('Temperature')
plt.title('Temperature Plot')

plt.savefig("temps.pdf")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
troops = pd.read_table("troops.txt", header=True)
cities = pd.read_table("cities.txt", header=True)
temps = pd.read_table("temps.txt", header=True, parse_dates=["date"])

# Create the first plot (Minard chart)
fig, ax = plt.subplots(figsize=(16, 4))

for _, row in troops.iterrows():
    ax.plot(row['long'], row['lat'], marker='o', markersize=row['survivors'] / 1000, color='grey' if row['direction'] == 'A' else 'red')

ax.scatter(cities['long'], cities['lat'], color='black')
for _, row in cities.iterrows():
    ax.text(row['long'], row['lat'], row['city'], ha='right', va='top', fontsize=8)

ax.set_xlim(24, 39)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Napoleon\'s March Minard Chart')

plt.savefig("march.pdf")

# Create the second plot (Temperature chart)
plt.figure(figsize=(16, 4))
sns.lineplot(x='long', y='temp', data=temps)
for _, row in temps.iterrows():
    plt.text(row['long'], row['temp'], f"{row['day']} {row['month']}", va='top', fontsize=8)
plt.xlim(24, 39)
plt.xlabel('Longitude')
plt.ylabel('Temperature')
plt.title('Temperature Plot')

plt.savefig("temps.pdf")

plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Read the data
troops = pd.read_table("troops.txt", header=True)
cities = pd.read_table("cities.txt", header=True)
temps = pd.read_table("temps.txt", header=True, parse_dates=["date"])

# Create the first plot (Minard chart)
minard_fig = px.scatter_geo(
    troops, 
    lat='lat', lon='long', 
    size='survivors', color='direction',
    text='city',
    projection='orthographic',
    title='Napoleon\'s March Minard Chart'
)
minard_fig.update_geos(
    showcoastlines=True, coastlinecolor="Black",
    showland=True, landcolor="white",
    showocean=True, oceancolor="lightblue"
)
minard_fig.add_trace(px.scatter_geo(cities, lat='lat', lon='long')['data'][0])
minard_fig.write_html("minard_chart.html")

# Create the second plot (Temperature chart)
temps_fig = px.line(
    temps, x='long', y='temp', 
    labels={'long': 'Longitude', 'temp': 'Temperature'},
    title='Temperature Plot'
)
temps_fig.add_trace(
    px.scatter(
        temps, x='long', y='temp', text=temps['day'] + ' ' + temps['month']
    )['data'][0]
)
temps_fig.write_html("temperature_plot.html")

# Display the plots
minard_fig.show()
temps_fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Read the data
troops = pd.read_table("troops.txt", header=True)
cities = pd.read_table("cities.txt", header=True)
temps = pd.read_table("temps.txt", header=True, parse_dates=["date"])

# Create the Minard chart with interactive features
minard_fig = px.scatter_geo(
    troops,
    lat='lat', lon='long',
    size='survivors', color='direction',
    text='city',
    projection='orthographic',
    title='Napoleon\'s March Minard Chart',
    opacity=0.7,
    template='plotly',
    hover_name='city',
    hover_data={'survivors': ':.1f'},
)
minard_fig.update_geos(
    showcoastlines=True, coastlinecolor="Black",
    showland=True, landcolor="white",
    showocean=True, oceancolor="lightblue"
)
minard_fig.add_trace(px.scatter_geo(cities, lat='lat', lon='long')['data'][0])
minard_fig.update_layout(geo=dict(showframe=False))
minard_fig.write_html("interactive_minard_chart.html")

# Create the Temperature chart with interactive features
temps_fig = px.line(
    temps, x='long', y='temp',
    labels={'long': 'Longitude', 'temp': 'Temperature'},
    title='Temperature Plot',
    template='plotly',
    line_shape='linear',
    hover_name='date',
    hover_data={'temp': ':.1f'},
)
temps_fig.add_trace(
    px.scatter(
        temps, x='long', y='temp', text=temps['day'] + ' ' + temps['month']
    )['data'][0]
)
temps_fig.write_html("interactive_temperature_plot.html")

# Display the plots
minard_fig.show()
temps_fig.show()
```
